```
import numpy as np

# ===============================
# Données (rendements mensuels %)
# ===============================
rendements_A = np.array([
    1.2, 0.8, -0.5, 1.5, 0.9, 1.1, 0.7, 1.3, 1.0, 0.6, 1.4, 0.8,
    1.1, 0.9, -0.3, 1.2, 1.0, 1.5, 0.8, 1.3, 0.9, 1.1, 1.2, 1.0
])

rendements_B = np.array([
    4.5, -2.1, 6.2, -3.5, 5.8, 7.1, -1.8, 4.9, 3.2, -4.2, 8.5, -2.7,
    5.1, 6.8, -3.1, 7.3, 4.5, -2.9, 6.7, 5.3, -3.8, 7.9, 4.2, 5.5
])

# ===============================
# Fonction statistiques
# ===============================
def stats_portefeuille(r):
    moyenne = np.mean(r)
    ecart_type = np.std(r, ddof=1)
    mediane = np.median(r)
    rendement_annuel = ((1 + moyenne/100)**12 - 1) * 100
    volatilite_annuelle = ecart_type * np.sqrt(12)

    return moyenne, ecart_type, mediane, rendement_annuel, volatilite_annuelle

# Résultats
stats_A = stats_portefeuille(rendements_A)
stats_B = stats_portefeuille(rendements_B)

print("PORTEFEUILLE A")
print(f"Moyenne mensuelle : {stats_A[0]:.2f}%")
print(f"Écart-type mensuel : {stats_A[1]:.2f}%")
print(f"Médiane : {stats_A[2]:.2f}%")
print(f"Rendement annualisé : {stats_A[3]:.2f}%")
print(f"Volatilité annualisée : {stats_A[4]:.2f}%")

print("\nPORTEFEUILLE B")
print(f"Moyenne mensuelle : {stats_B[0]:.2f}%")
print(f"Écart-type mensuel : {stats_B[1]:.2f}%")
print(f"Médiane : {stats_B[2]:.2f}%")
print(f"Rendement annualisé : {stats_B[3]:.2f}%")
print(f"Volatilité annualisée : {stats_B[4]:.2f}%")
```
